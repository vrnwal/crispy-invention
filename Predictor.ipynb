# Poisson Formula

```python
def factorial(x):
    if x == 0:
        return 1
    if x == 1:
        return 1
    return x * factorial(x - 1)
```

```python
def poisson(goals, expected):
    return e ** (-1 * expected) * (expected ** goals ) / factorial(goals) * 100
```

# Standardized Outcome Generator

For a given match in a season, we will predict the outcome given the matches that have occured thus far in the season

```python
def query():
    home = input("Please select the Home team: ")
    away = input("Please select the Away team: ")
    print(type(home), (type(away)))
    return [home, away]
```

```python
def predict(home, away, season):
    s = pd.read_csv(season)
    reduced_s = s.drop(s.columns[38:], axis = 1)
    # Select the matchday to predicy only using the games played so far in the season
#     tble = reduced_s.copy()
#     tble = tble.where(tble['HomeTeam'] == home)
#     tble = tble.where(tble['AwayTeam'] == away)
#     tble = tble.sort_values('HomeTeam')
    total_games = reduced_s.shape[0]
    games_played = total_games / 20
    average_attack_home = sum(reduced_s['FTHG']) / total_games
    average_attack_away = sum(reduced_s['FTAG']) / total_games
    average_defense_home, average_defense_away = average_attack_away, average_attack_home
    
    # Calculate home teams goal scoring chances
    home_team = goal_counter(home, True, season)
    
    home_goals = home_team[0] / games_played
    home_conceded = home_team[1] / games_played
    
    home_goals_scored = home_goals / average_attack_home 
    home_goals_conceded = home_conceded / average_defense_home
    
    # Calculate away teams goal scoring chances
    away_team = goal_counter(away, False, season)
    
    away_goals = away_team[0] / games_played
    away_conceded = away_team[1] / games_played
    
    away_goals_scored = away_goals / average_attack_away
    away_goals_conceded = away_conceded / average_defense_home
    
    # Predicting goals scored 
    home_prediction = home_goals_scored * away_goals_conceded * average_attack_home
    away_prediction = away_goals_scored * home_goals_conceded * average_defense_home

    # Now we predict most likely number of goals for each team using Poisson
    # Home 
    home_range = []
    for i in range(0,7):
        home_range.append(poisson(i, home_prediction))
    # Away
    away_range = []
    for i in range(0,7):
        away_range.append(poisson(i, away_prediction))
    highest_home = max(home_range)
    highest_away = max(away_range)
    
    return home_range.index(highest_home), away_range.index(highest_away)
    
```

# Modules

```python
def goal_counter(team, home_or_away, season):
    s = pd.read_csv(season)
    if home_or_away == True:
        s = s.where(s['HomeTeam'] == team)
        s = s.sort_values('HomeTeam').iloc[:19]
        
        goals_scored = sum(s['FTHG'])
        goals_conceded = sum(s['FTAG'])
    elif home_or_away == False:
        s = s.where(s['AwayTeam'] == team)
        s = s.sort_values('AwayTeam').iloc[:19]
        
        goals_scored = sum(s['FTAG'])
        goals_conceded = sum(s['FTHG'])
    
    return [goals_scored, goals_conceded]
```
